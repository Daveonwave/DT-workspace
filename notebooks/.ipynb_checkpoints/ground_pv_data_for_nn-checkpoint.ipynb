```python
import torch
import pandas as pd
import os
from pathlib import Path
from sklearn.model_selection import train_test_split
import numpy as np
import tqdm
import copy
from matplotlib import pyplot as plt
```

```python
ground = pd.read_csv("../data/output/nn/simulation/PV_profile/2024_03_01-16_58_04/ground.csv")
sim = pd.read_csv("../data/output/nn/simulation/PV_profile/2024_03_01-16_58_04/dataset.csv")
temp_1 = ground['temperature'].tolist()
temp_1.insert(0, 295.15)
ground['lag_1'] = temp_1[:-1]
ground['lag_2'] = ground['lag_1'].shift(1)
ground['lag_3'] = ground['lag_2'].shift(1)
ground['lag_4'] = ground['lag_3'].shift(1)
ground['lag_5'] = ground['lag_4'].shift(1)
ground['rolling_mean'] = ground[[col for col in ground.columns if 'lag' in col]].mean(axis=1)
ground['t_amb'] = [295.15] * len(ground)

ground['heat'] = sim['heat'][1:].tolist()
```

```python
ground = ground.dropna()
```

```python
from sklearn.model_selection import train_test_split
from statsmodels.tsa.statespace.sarimax import SARIMAX

def SARIMAX_model(train,test):
 
    # fit model
    model = SARIMAX(train.drop(['current', 'temp-1', 't_amb', 'heat'], axis=1), exog=train[['current', 'temp-1', 't_amb', 'heat']], order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
    model_fit = model.fit(disp=False)
    
    # make prediction
    yhat = model_fit.predict(len(train), len(train) + len(test) - 1, exog=test[['current', 'temp-1', 't_amb', 'heat']].values)
    res=pd.DataFrame({"Pred":yhat, "Act": test["temperature"].values, "Exog":test[['current', 'temp-1', 't_amb', 'heat']].values})
    return res
```

```python
from statsmodels.tsa.arima.model import ARIMA

def ARIMA_model(train,test):
    # fit model
    model = ARIMA(train['temperature'], order=(1, 1, 1))
    model_fit = model.fit()
    # make prediction
    yhat = model_fit.predict(len(train), len(train) + len(test) - 1, typ='levels')
    res = pd.DataFrame({"Pred": yhat, "Act": test["temperature"].values})
    return res
```

```python
#df_train = df_train.drop('voltage', axis=1)
train, test = train_test_split(df_train, test_size=0.2)

df_ret = ARIMA_model(train, test)
plt.plot(range(len(test['temperature'])), test['temperature'], label="ground")
plt.plot(range(len(test['temperature'])), df_ret['Pred'], label='prediction')
```
